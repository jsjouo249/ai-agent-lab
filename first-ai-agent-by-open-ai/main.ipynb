```python
from dotenv import load_dotenv
import openai

load_dotenv()

client = openai.OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
      {"role": "user", "content": "What is the capital of the greece?"}
    ],
)

print(response)
```

```python
message = response.choices[0].message.content

message
```
